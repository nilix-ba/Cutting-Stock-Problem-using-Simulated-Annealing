```python
import pandas as pd
import numpy as np
import os
from os import listdir
from os.path import isfile, join
import random
```

reads the input and pre-process it

```python
def read(path):
    
    global length
    global req
    global size
    
    file = open(path)
    inp = file.read()
    file.close()
    
    row = inp.split('\n')
    #print(row[0])
    
    helper = row[0].split(':')
    #row[0]
    #print(helper)
    
    length = int(helper[1])
    #print(length)
    
    helper.clear()
    helper = row[3].split(',')
    #row[3]
    
    req = list(map(int, helper))
    size = len(req)
    #print(req)
```

the first permutation

```python
def perm(req):
    
    random.shuffle(req)
    #print(req)
    
    return req
```

neighbourhood function 1: simply swaps two of the indexes choosen randomly

```python
def N1(ls):
    
    global size
    
    i1 = np.random.randint(0, size-1)
    i2 = np.random.randint(i1 + 1, size)
    
    ls[i1], ls[i2] = ls[i2], ls[i1]
    
    return ls
```

neighbourhood function 2: simply swaps three of the indexes choosen randomly

```python
def N2(ls):
    
    global size
    
    i1 = np.random.randint(0, size-1)
    i2 = np.random.randint(i1 + 1, size)
    i3 = np.random.randint(0, size)
    
    ls[i1], ls[i2] = ls[i2], ls[i1]
    ls[i2], ls[i3] = ls[i3], ls[i2]
    
    return ls
```

the fitness funtion which will calculate how much stock material a single permutation that it's given will take

```python
def fitness(ls):
    
    global length
    global size
    
    i = 1
    total = ls[0]
#     print(total)
    cost = 0
    
    while i != size:
        
        total += ls[i] #adds on the requested sizes 
        
#         print("ls = ", end= '')
#         print(ls[i])
#         print("total = ", end= '')
#         print(total)
#         print("i= ", end= '')
#         print(i)
        
        if total > length: #if the sum of requested sizes exceeded the material's lenght we will add 1 to the total
                          #materials used
            
            total = 0
            cost += 1
            i -= 1
            
#             print("cost = >  ", end= '')
#             print(cost)
#             print("i = > ", end= '')
#             print(i)
#             print("total = > ", end= '')
#             print(total)
            
        i += 1
    
    cost += 1
    
        
    return cost
```

calculated the probability of replacing a best answer with a worse one

```python
def probability(diff): 
    
    global T
    
    tmp = np.exp(-(diff/T))
    p = random.uniform(0, 1)
    
    if p <= tmp:
        
        return True
    
    return False
```

main function for processing the data

# inpt 1

```python
def main():
    
    global length
    global req
    global size
    global T
    
    read("input1.stock")
    perm(req)
    
    ans = req[:]
    cost = fitness(ans)
    print(cost)
    
    while cost > 55 : #while the desired answer isnt found do
           
#         print('cost in while =', end= ' ')
#         print(cost, end = '\n')
#         print('answer in while =', end= ' ')
#         print(ans)
        
        N = ans[:]
        cost_N = 100000000
        
        for i in range(10): #make 10 neighbours from the first neighbour func and choose the best one
            
            n1 = N1(N)
            c1 = fitness(n1)
            
            if c1 < cost_N:
                
                N = n1[:]
                cost_N = c1
                
        for i in range(10): #make 10 neighbours from the second neighbour func and choose the best one
            
            n2 = N2(N)
            c2 = fitness(n2)
            
            if c2 < cost_N:
                
                N = n2[:]
                cost_N = c2
        
        if cost_N < cost: #if the neighbour was better than the best answer 100% replacement
            
            ans = N[:]
            cost = cost_N
            
            print('cost b=', end= ' ')
            print(cost, end = '\n')
            print('answer b=', end= ' ')
            print(ans)
            
        else: #if it is worse then replace it with a probability
            
            diff = cost_N - cost
            
            if diff != 0:
                
                flag = probability(diff)
            
                if flag:

                    ans = N[:]
                    cost = cost_N
                    print('cost c=', end= ' ')
                    print(cost, end = '\n')
                    print('answer c=', end= ' ')
                    print(ans)
                                        
        T = 0.8*T
                    
        
    print('cost =', end= ' ')
    print(cost, end = '\n')
    print('answer =', end= ' ')
    print(ans)
    
```

```python
global length
global req
global size
global T

T = 30

main()
```

So far T = 30 gave the best and fastest results
